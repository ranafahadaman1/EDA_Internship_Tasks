#### Solution:

##### This code calculates the correlation matrix for the DataFrame 'menuSelectedDf' and then plots it as a heatmap using Seaborn. The heatmap visualizes the correlation between different attributes, with higher values indicating stronger correlations.

### Task6 
#####  Draw Stripplot for each category against the following attributes {'Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'.Value)'.

#### Solution:

##### This code defines a list of attributes and sets up a figure with multiple subplots to display strip plots for each attribute by category. It iterates through each attribute, creates a stripplot for each category against the current attribute using Seaborn, and adjusts the layout to provide more spacing between subplots for better visualization.

```python
# Define the list of attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 
              'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 
              'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
              'Iron (% Daily Value)']

# Set up the figure and axes
fig, axes = plt.subplots(nrows=len(attributes), ncols=1, figsize=(8, 80))

# Iterate through each attribute
for i, attribute in enumerate(attributes):
    # Create a stripplot for each category against the current attribute
    sns.stripplot(x='Category', y=attribute, hue='Category', data=menuDf, ax=axes[i], dodge=False)
    axes[i].set_title(f"{attribute}")
    #axes[i].set_xlabel("Category", fontsize=12)
    #axes[i].set_ylabel(attribute, fontsize=12)
    axes[i].tick_params(axis='x', rotation=70)  # Rotate x-axis labels
    axes[i].grid(axis='y', linestyle='-')  # Add gridlines parallel to x-axis
    
# Adjust layout with more spacing between subplots
plt.subplots_adjust(hspace=0.75)

plt.show()
```

### Task7
#####  Draw a horizontal bar graph for items in each category against the calories.


#### Solution:

##### This code creates subplots for each unique category in the DataFrame 'menuDf'. For each category, it filters the data, sorts it by calories in ascending order, and creates a horizontal bar plot to visualize the calories in menu items. Each subplot represents a category, and items are grouped by their calorie counts.


```python
# Get the unique categories
categories = menuDf['Category'].unique()

# Set up subplots
fig, axes = plt.subplots(nrows=len(categories), figsize=(15, 12 * len(categories)))

# Loop through each category
for i, category in enumerate(categories):
    # Filter the data for the current category
    categoryData = menuDf[menuDf['Category'] == category]
    
    # Sort the data by calories in ascending order
    categoryDataSorted = categoryData.sort_values(by='Calories', ascending=True)
    
    # Create a horizontal bar plot for items in the current category against calories
    sns.barplot(x='Calories', y='Item', data=categoryDataSorted, ax=axes[i], hue='Item')
    
    # Set labels and title for the current subplot
    axes[i].set_xlabel('Calories')
    axes[i].set_ylabel('Menu Items')
    axes[i].set_title(f'Calories in {category} Menu Items')
    axes[i].grid(True, linewidth=0.25)

# Adjust layout
plt.tight_layout()
plt.show()
```
